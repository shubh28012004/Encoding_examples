1) Method 1 :- .cat method

```python
import pandas as pd
import numpy as np
```

```python
#creating initial dataframe
bridge_types = ('Arch' , 'Bean' ,'Truss' , 'Cantilever' , 'Tied Arch' , 'Suspension' , 'Cable')
bridge_df = pd.DataFrame(bridge_types , columns=['Bridge_types'])
```

```python
#converting type of columns tp category
bridge_df['Bridge_types'] = bridge_df['Bridge_types'].astype('category')
```

```python
#Assigning numericalical values and storing in another column
bridge_df['Bridge_types_Cat'] = bridge_df['Bridge_types'].cat.codes
bridge_df
```

Method 2 :- using sci-kit library approach

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#creating initial dataframe
bridge_types = ('Arch' , 'Bean' ,'Truss' , 'Cantilever' , 'Tied Arch' , 'Suspension' , 'Cable')
bridge_df = pd.DataFrame(bridge_types , columns=['Bridge_types'])
```

```python
#Creating instance of label encoder
labelencoder = LabelEncoder()
```

```python
#Assigning numerical values and storing it to the columns
bridge_df['Bridge_types_Cat'] = labelencoder.fit_transform(bridge_df['Bridge_types'])
```

```python
bridge_df
```

one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#creating instance of onehot encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
#passinf bridge-type-cat column (label encoded values of bridge_type )
enc_df = pd.DataFrame(enc.fit_transform(bridge_df[['Bridge_types_Cat']]).toarray())
```

```python
#merge with main df bridge_df on any key values
bridge_df = bridge_df.join(enc_df)
bridge_df
```

Using dummies values approach

```python
#creating initial dataframe
bridge_types = ('Arch' , 'Bean' ,'Truss' , 'Cantilever' , 'Tied Arch' , 'Suspension' , 'Cable')
bridge_df = pd.DataFrame(bridge_types , columns=['Bridge_Types'])
```

```python
#generate binary values using get_dummies
dum_df = pd.get_dummies(bridge_df,columns=["Bridge_Types"],prefix = ["Type_is"])
```

```python
# merge with main df bridge_df on key values
bridge_df = bridge_df.join(dum_df)
```
